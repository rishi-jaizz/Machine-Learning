```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("student_clustering.csv")
print("The shape of data is",df.shape)
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'],df['iq'])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i) #n_clusters specifies the number of clusters
    km.fit_predict(df)
    #fit_predict does the functionality of both fitting and prediction
    wcss.append(km.inertia_)
    # calculates the within cluster sum of square of distance between point and centroid
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
x=df.iloc[:,:].values
#x is a numpy array in 2D
km=KMeans(n_clusters=4)
y_means=km.fit_predict(x)
#y means stores the values of cluster assigned to each data points in dataframes
```

```python
x[y_means == 3,1]
```

```python
plt.scatter(x[y_means == 0,0],x[y_means == 0,1],color='blue')
plt.scatter(x[y_means == 1,0],x[y_means == 1,1],color='red')
plt.scatter(x[y_means == 2,0],x[y_means == 2,1],color='green')
plt.scatter(x[y_means == 3,0],x[y_means == 3,1],color='yellow')
```
